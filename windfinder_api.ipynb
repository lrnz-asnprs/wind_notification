```python
import requests

variables = 'WindSpeed'
lat = '55.5633'
lon = '12.5636'

# Return timeseries data
url = f"https://api.fcoo.dk/metoc/v2/data/timeseries?variables=WindSpeed,Wind&lat={lat}&lon={lon}"

# Return plot for given location with timeseries
# url = f"https://metoc.fcoo.dk/webapi/plot/timeseries_profile?p=WindSpeed&x={lon}&y={lat}&nx=800&ny=800"

# Get list of available forecaset parameters
# url = f"https://api.fcoo.dk/metoc/v2/info/parameters"

# Make the API call
response = requests.get(url)

# Check the response status code
if response.status_code == 200:
    # API call successful
    data = response.json()
    # Process the response data as needed
    print(data)
else:
    # API call failed
    print("Error:", response.status_code)
```

```python
# Data
windspeed_data = data['WindSpeed']['windspeed']['data']
windspeed_timestamps = data['WindSpeed']['windspeed']['time']
```

```python
import pandas as pd

df = pd.DataFrame({'timestamp':windspeed_timestamps, 'windspeed': windspeed_data, 'lat': lat, 'lon': lon})

df['timestamp'] = pd.to_datetime(df['timestamp'])

df.head(50)
```

```python
def extract_date_info(date):
    hour = date.hour
    day = date.strftime('%A')
    month = date.month
    return pd.Series([hour, day, month])

df[['hour', 'day', 'month']] = df['timestamp'].apply(extract_date_info)
```

```python
df['rolling_3h_average'] = df['windspeed'].rolling(window=3).mean().values
```
